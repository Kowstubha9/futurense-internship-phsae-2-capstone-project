# Importing Libraries

```python
import mysql.connector
from mysql.connector import Error
import pandas as pd
```

# Connecting to MySQL Server

```python
def create_server_connection(host_name, user_name, user_password,port_no):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            password=user_password,
            port=port_no
        )
        print("MySQL server connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection

connection=create_server_connection("localhost","root","Mysqlroot1@3$",3307)
```

# Creating a Database

```python
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully")
    except Error as err:
        print(f"Error: '{err}'")

query= "CREATE DATABASE sports"
create_database(connection,query)
```

# Connecting to MySQL Database

```python
def create_db_connection(host_name, user_name, user_password, db_name, port_no):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name,
            port=port_no
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection

connection=create_db_connection("localhost","root","Mysqlroot1@3$","sports",3307)
```

# Creating Query Execution Function

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

# Creation of Tables

## Player_info

```python
query="CREATE TABLE Player_info (SerialID INT PRIMARY KEY, Player VARCHAR(60), Team VARCHAR(60), Position VARCHAR(60), Season INT);"
execute_query(connection, query)
```

## Player_demographics

```python
query="CREATE TABLE Player_demographics (SerialID INT, Age INT NOT NULL, Height FLOAT, Weight FLOAT, FOREIGN KEY(SerialID) REFERENCES Player_info(SerialID));"
execute_query(connection, query)
```

## Player_stats

```python
query="CREATE TABLE Player_stats (SerialID INT, Goals INT, Assists INT, YellowCards INT, RedCards INT, PassCompletionRate FLOAT, DistanceCovered FLOAT, Sprints INT, ShotsOnTarget INT, TacklesWon INT, CleanSheets INT, FOREIGN KEY(SerialID) REFERENCES Player_info(SerialID));"
execute_query(connection, query)
```

## Player_training_info

```python
query="CREATE TABLE Player_training_info (SerialID INT, TrainingHours FLOAT, EffectiveTraining FLOAT, FOREIGN KEY(SerialID) REFERENCES Player_info(SerialID));"
execute_query(connection, query)
```

## Player_performance

```python
query="CREATE TABLE Player_performance (SerialID INT, PressurePerformanceImpact FLOAT, MatchPressure INT, FOREIGN KEY(SerialID) REFERENCES Player_info(SerialID));"
execute_query(connection, query)
```

## Player_health

```python
query="CREATE TABLE Player_health (SerialID INT, InjuryHistory INT, PlayerFatigue FLOAT, FatigueInjuryCorrelation FLOAT, FOREIGN KEY(SerialID) REFERENCES Player_info(SerialID));"
execute_query(connection, query)
```

# Retrieving Data

```python
def read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")
```

```python
query="select SerialId,Player,Team from Player_info limit 3;"
results = read_query(connection, query)
for result in results:
    print(result)
```
